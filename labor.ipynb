```python
import pandas as pd
df = pd.read_csv('http://www.akyokus.com/dataset/labor.csv')
```

```python
df.describe()
```

```python
df.info()
```

```python
df.head(6)
```

```python
df.replace('?', 0, inplace=True)
df.replace('none', 0,  inplace=True)
```

```python
df['class'].value_counts()
```

```python
df['working-hours'].value_counts().plot(kind='bar')
```

```python
from sklearn.preprocessing import OrdinalEncoder
df=df.astype(str)
encoder = OrdinalEncoder().set_output(transform="pandas")
encoder = encoder.fit(df)
encoded_df = encoder.transform(df)
encoded_df
```

```python
y = encoded_df.pop('class')
```

```python
X = encoded_df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=43) # 70% eiitim and 30% test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_clf = DecisionTreeClassifier()
```

```python
dt_clf = dt_clf.fit(X_train,y_train)
```

```python
y_pred1 = dt_clf.predict(X_test)
```

```python
from sklearn.metrics import precision_score, \
    recall_score, confusion_matrix, classification_report, \
    accuracy_score, f1_score

print( 'Accuracy:', accuracy_score(y_test,y_pred1))
print( 'F1 score:', f1_score(y_test, y_pred1))
print( 'Recall:', recall_score(y_test, y_pred1))
print( 'Precision:', precision_score(y_test, y_pred1))
print( '\n clasification report:\n', classification_report(y_test,y_pred1))
print( '\n confussion matrix:\n',confusion_matrix(y_test, y_pred1))
```
